```python
# Importamos las bibliotecas necesarias 
import pandas as pd                  # Para trabajar con DataFrames ( tablas de datos )
import numpy  as np                  # Para algebra lineal ( funciones 'vectoriales' )
import matplotlib.pyplot as plt      # Para graficar 

# Nota: En caso de trabajar en Python 2.7.x, es recomendable colocar la siguiente instrucción:
# %matplotlib inline                   # Instrucción para indicar que queremos las graficas en el Notebook
```

```python
datos_1 = pd.read_csv("datos_ajuste_1.csv") # Importamos el archivo 1
datos_2 = pd.read_csv("datos_ajuste_2.csv") # Importamos el archivo 2
datos_3 = pd.read_csv("datos_ajuste_3.csv") # Importamos el archivo 3
datos_4 = pd.read_csv("datos_ajuste_4.csv") # Importamos el archivo 4

# print( datos_1.head() )
# print( datos_2.head() )
# print( datos_3.head() )
# print( datos_4.head() )
```

```python
X, Y = datos_1["X"], datos_1["Y"] 

# Realizamos un grafico de dispersión ( Scatter-Plot )
plt.figure( figsize=(16,6) )                           # Indicamos el tamaño de la gráfica
plt.title("Scatter Plot")                              # Añadimos el titulo
plt.scatter( X, Y , label = "Etiqueta de la serie" )   # Añadimos la 'capa' de puntos 
plt.xlabel("Eje X")                                    # Modificamos la etiqueta del eje X
plt.ylabel("Eje Y")                                    # Modificamos la etiqueta del eje Y
plt.grid()                                             # Añadimos una cuadricula/rendija 
plt.legend()                                           # Añadimos un cuadro con la leyenda
plt.show()                                             # Indicamos que queremos mostrar la gráfica
```

```python
X, Y = datos_1["X"], datos_1["Y"] 

# Realizamos un grafico de dispersión ( Scatter-Plot )
plt.figure( figsize=(16,6) )                           # Indicamos el tamaño de la gráfica
plt.title("Scatter Plot")                              # Añadimos el titulo
plt.scatter( X, Y , label = "Etiqueta de la serie" )   # Añadimos la 'capa' de puntos 
# El modelo que ajustamos fue:
# y = b0 + b1*x
plt.plot(    X, -15 + np.sqrt(2)*X , label = "Regresión", color="red" )   # Añadimos la 'capa' de puntos 
plt.xlabel("Eje X")                                    # Modificamos la etiqueta del eje X
plt.ylabel("Eje Y")                                    # Modificamos la etiqueta del eje Y
plt.grid()                                             # Añadimos una cuadricula/rendija 
plt.legend()                                           # Añadimos un cuadro con la leyenda
plt.show()                                             # Indicamos que queremos mostrar la gráfica
```

```python
X, Y = datos_2["X"], datos_2["Y"] 

# Realizamos un grafico de dispersión ( Scatter-Plot )
plt.figure( figsize=(16,6) )                           # Indicamos el tamaño de la gráfica
plt.title("Scatter Plot")                              # Añadimos el titulo
plt.scatter( X, Y , label = "Etiqueta de la serie" )   # Añadimos la 'capa' de puntos 
# El modelo que ajustamos fue:
# y = b0 + b1*x + b2*x^2 
plt.plot( X, -15 + np.sqrt(2)*X - 0.2*X**2 , label = "Regresión", color="red" )   # Añadimos la 'capa' de puntos 
plt.xlabel("Eje X")                                    # Modificamos la etiqueta del eje X
plt.ylabel("Eje Y")                                    # Modificamos la etiqueta del eje Y
plt.grid()                                             # Añadimos una cuadricula/rendija 
plt.legend()                                           # Añadimos un cuadro con la leyenda
plt.show()                                             # Indicamos que queremos mostrar la gráfica
```

```python
X, Y = datos_3["X"], datos_3["Y"] 

# Realizamos un grafico de dispersión ( Scatter-Plot )
plt.figure( figsize=(16,6) )                           # Indicamos el tamaño de la gráfica
plt.title("Scatter Plot")                              # Añadimos el titulo
plt.scatter( X, Y , label = "Etiqueta de la serie" )   # Añadimos la 'capa' de puntos 
# El modelo que ajustamos fue:
# y = b0 + b1*x + b2*sin(2*pi*x/48) + b3*cos(2*pi*x/48)
plt.plot( X, 2 + 0.02*X + np.sqrt(2)*np.sin(2*np.pi*X/48) + np.sqrt(3)*np.cos(2*np.pi*X/48) , label = "Regresión", color="red" )   # Añadimos la 'capa' de puntos 
plt.xlabel("Eje X")                                    # Modificamos la etiqueta del eje X
plt.ylabel("Eje Y")                                    # Modificamos la etiqueta del eje Y
plt.grid()                                             # Añadimos una cuadricula/rendija 
plt.legend()                                           # Añadimos un cuadro con la leyenda
plt.show()                                             # Indicamos que queremos mostrar la gráfica
```
