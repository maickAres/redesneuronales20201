**¿Qué es NumPy?**

- Una biblioteca para Python: ndarray + ufunc
- Los arrays multidimensionales (ndarray) nos permiten almacenar datos de manera estructurada
- Las funciones universales (ufunc) nos permiten operar con esos datos de manera eficiente

Python está organizado en módulos, que son archivos con extensión .py que contienen funciones, variables y otros objetos, y paquetes, que son conjuntos de módulos. Cuando queremos utilizar objetos que están definidos en un módulo tenemos que importarlo, y una vez que lo hemos hecho podemos usar el operador . para ir descendiendo en la jerarquía de paquetes y acceder al objeto que necesitamos. Por ejemplo, de esta manera importamos NumPy:

```python
import numpy as np      # esta siempre es la conversion para numpy
```

para acceder a la funcion eigvals primero hay que llamar el paquete numpy y despues hay que llamar al paquete linalg y en este encontraremos la funcion

**¿Qué es exactamente un array?**

Un array de NumPy es una colección de N elementos, igual que una secuencia de Python (por ejemplo, una lista). 

Los arrays de NumPy son homogéneos, es decir, todos sus elementos son del mismo tipo. Si le pasamos a np.array una secuencia con objetos de tipos diferentes, promocionará todos al tipo con más información. Para acceder al tipo del array, podemos usar el atributo dtype.

**¿Por qué? Eficiencia**

- Los bucles son costosos
- Eliminar bucles: vectorizar operaciones
- Los bucles se ejecutan en Python, las operaciones vectorizadas en C
- Las operaciones entre arrays de NumPy se realizan elemento a elemento

*Ejemplo:*

Sea $A, B, C \in M_{N \times K}$, deseamos obtener la suma de dos matrices.

Es decir, $A = B + C$. Por lo que debemos ejecutar la siguiente operación:

$$ a_{ij} = b_{ij} + c_{ij} \qquad \qquad \forall i \in \{1,2,\ldots,N\}, \ j \in \{1,2,\ldots,K\}$$

```python
N, K = 100, 100                           #con esto creamos tres matrices de NxK de manera aleatoria 
A = np.empty(10000).reshape(N, K)
B = np.random.rand(10000).reshape(N, K)
C = np.random.rand(10000).reshape(N, K)
```

```python
%%timeit # con este comando mostramos el tiempo que tarda en realizarse la operacion en la celda 
for i in range(N):
    for j in range(K):
        A[i, j] = B[i, j] + C[i, j]
```

```python
%%timeit # en esta celda vemos que vectorizar las operaciones nos genera un menor tiempo de ejecucion 
A = B + C
```

```python
A=np.matrix(A) #esta funcion transforma una lista o dataframe en una matrix 
B=np.matrix(B) 
C=np.matrix(C)
```

```python
%%timeit 
Z=B+C
```

```python
B=np.array([[1,2],[5,3]])
C=np.array([[2,2],[1,3]])
```

```python
D=B*C # vemos que la multiplicacion de arrays se realiza entrada a entrada 
D
```

**¿Qué es matplotlib?**

- Estándar de facto para visualización en Python
- Pretende ser similar a las funciones de visualización de MATLAB
- Diferentes formas de usarla: interfaz pyplot y orientada a objetos

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot([0, 0.1, 0.2, 0.5,0.6], [1, -1, 0, 3, -1])
plt.show()
```

La tarea más habitual a la hora de trabajar con matplotlib es representar una función.

Lo que tendremos que hacer es definir un dominio y evaluarla en dicho dominio. Por ejemplo:

$$f(x) = e^{-x^6}$$

```python
def f(x):
    return np.exp(-x ** 6)
```

```python
x = np.linspace(-10, 20, num=100) # esta funcion crea una serie de puntos equidistantes la cantidad de segmentos que se crearan 
                               # esta delimitada por num 
```

```python
plt.plot(x, f(x), label="Función $e^{-x^6}$ ")
plt.xlabel("Eje $x$")
plt.ylabel("$f(x)$") #Podemos añadir etiquetas, y escribir LaTeX en ellas. Tan solo hay que encerrarlo entre signos de dólar
plt.legend()
plt.title("Función $f(x) = e^{-x^6}$")
plt.show()  
```

```python
#aqui creamos una grafica de dispersión 
N = 50
x = np.random.randn(N)
y = np.random.randn(N)

plt.scatter(x, y)
plt.show()
```

Con `s` y `c` podemos modificar el tamaño y el color respectivamente. Para el color, a cada valor numérico se le asigna un color a través de un mapa de colores; ese mapa se puede cambiar con el argumento cmap. Esa correspondencia se puede visualizar llamando a la función colorbar.

```python
s = 50 + 50 * np.random.randn(N)
c = np.random.randn(N)

plt.scatter(x, y, s=s, c=c, cmap=plt.cm.Reds)
plt.colorbar()
plt.show()
```

```python
x = np.linspace(-10,10,100) # este es un ejemplo de como ocupar la interfaz orientada a objetos  
y = np.exp(x)

fig, axes = plt.subplots()

axes.plot(x, y, 'ro', label="Función F(x)")
axes.set_xlim(-7, 10)
axes.set_ylim(-7, 100)

fig.savefig("grafica1.png")  # Con este comando podemos exportar la gráfica a un archivo en nuestro ordenador
```

```python
fig, axes = plt.subplots(1, 2, sharey= True) #Con Sharey undicamos si quereos que comparta el eje y o no 

axes[0].plot(x, y, color="blue")
axes[0].set_xlabel("Eje x")

axes[1].plot(x, -y, 'r')
axes[1].set_xlabel("Eje x ")

plt.show()
```

# **Operación común con DataFrames :: Append**

```python
transacciones_1 = pd.read_excel(io="datasets/DATOS.xlsx",sheet_name="TRANSACTIONS_PART1")
transacciones_1.head()
```

```python
transacciones_2 = pd.read_excel(io="datasets/DATOS.xlsx",sheet_name="TRANSACTIONS_PART2")
transacciones_2.head()
```

```python
transacciones_final = transacciones_1.append( transacciones_2 )
transacciones_final
```

```python
# tambien podemos usar la funcion concat
pd.concat( [transacciones_1, transacciones_2] )
```

**Operación común con DataFrames :: Merge**

```python
cat_empleados = pd.read_excel(io="datasets/DATOS.xlsx",sheet_name="CAT_EMPLEADOS")
cat_empleados.head()
```

```python
transacciones_3 = pd.merge(left = transacciones_1, right = cat_empleados, how = "inner", left_on = "EMPLOYEE_ID", right_on = "EMPLOYEE_ID")
transacciones_3
```

Muchos utilizar la función `concat`, pero agrega un parámetro más.

Notemos que NO está haciendo el cruce correctamente, pues el cruce que realiza lo hace con base en el `index`, no en una columna.

En este caso no podemos realizar la operación, debido a que la cardinalidad de los indices no son iguales.


```python
pd.concat( [transacciones_1, cat_empleados], axis = 1 )
```

**Operación común con DataFrames :: Pivot**

```python
transacciones_enero = transacciones_1.loc[ transacciones_1.PERIOD == "2019M01" , :]# se tomara solo los datos de enero para evitar problemas
transacciones_enero.tail()
```

```python
transacciones_enero.pivot(index = "EMPLOYEE_ID", columns="PRODUCT_ID", values="SALE" )
```

```python
# El método Pivot_Table es más completo que el anterior
transacciones_pivot = transacciones_enero.pivot_table(index = "EMPLOYEE_ID", columns="PRODUCT_ID", values="SALE", \
                                                      aggfunc='mean', fill_value=0, margins=True, dropna=False )
transacciones_pivot
```

**Operación común con DataFrames :: Melt (UnPivot)**

```python
pd.melt( transacciones_pivot, id_vars = "EMPLOYEE_ID" , value_vars = ["PROD-0001","PROD-0002","PROD-0003","PROD-0004","PROD-0005", \
   "PROD-0006","PROD-0007","PROD-0008","PROD-0009","PROD-0010"], var_name="ID_OF_THE_PRODUCT", value_name="SALES_OF_THE_PRODUCT" )
```
