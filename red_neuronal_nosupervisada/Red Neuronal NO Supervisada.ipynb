## modelo de red neuronal competitiva

1.- sea $X=(x1,x2,...,xl)$ un conjunto de de vectores de entrada(Normalizados).

2.-generar los vectores de  pesos $w1, w2, ..., wk$ de manera aleatoria y normalizarlos

3.-seleccionar aleatoriamente $xj\in X$, realizamos la siguiente operacion: $xj*wi$ para $i=1,...,k$

4.-seleccionar $wm$ talque $wm*xj>= wi*xj$ para $i=1,...,k$

5.- sustituir $wm$ por $wm+xj$ y normaliza

```python
z=pd.read_csv('5_clusters_2D.txt',header=None,sep=";")
```

```python
z=np.array(z)
```

```python
z
```

```python
k=3
```

```python
def normalizar(z1):
    for i in range(len(z1)):
        z1[i]=z1[i]/np.linalg.norm(z1[i],2)
    return z1
```

```python
def producto_punto(x,w):
    return sum(x[i]*w[i] for i in range(len(x)))
```

```python
w={}
for j in range(k):
    w[j]=np.random.rand(2)
```

```python
def aprendizaje(z1,w):
    x=np.zeros(k)
    for i in range(k):
        x[i]=producto_punto(z1,w[i])
    y=np.max(x)
    z=[x[l]==y for l in range(k) ]
    for j in range(k):
        if z[j]==True:
            w[j]=(w[j]+z1)/np.linalg.norm(w[j]+z1,2)
```

```python
def cluster_red(z,w,it):
    l=0
    w=normalizar(w)
    z1=normalizar(z)
    while  l<it :
        a=np.random.choice(len(z))
        aprendizaje(z1[a],w)
        l+=1
    return w
```

```python
w1=cluster_red(z,w,1000)
```

```python
def clase(z1):
    x=np.zeros(k)
    for i in range(k):
        x[i]=producto_punto(z1,w[i])
    y=np.max(x)
    z=[x[l]==y for l in range(k) ]
    for j in range(k):
        if z[j]==True:
            m=j
    return m
        
```

```python
def red_neuronal(z,w):
    z1=normalizar(z)
    x=np.zeros(len(z))
    for i in range(len(z)):
        x[i]=clase(z1[i])
    return x
```

```python
red_neuronal(z,w1)
```
