```python
# Importamos las bibliotecas necesarias 
import numpy  as np                  # Para algebra lineal ( funciones 'vectoriales' )
import matplotlib.pyplot as plt      # Para graficar 
# 
# Nota: En caso de trabajar en Python 2.7.x, es recomendable colocar la siguiente instrucci贸n:
# %matplotlib inline                   # Instrucci贸n para indicar que queremos las graficas en el Notebook
```

```python
# Creamos 2 arreglos a los que aplicaremos la regresi贸n lineal
x = np.array( [30, 28, 32, 25, 25, 25, 22, 24, 30, 40] )
y = np.array( [25, 30, 27, 40, 42, 40, 50, 45, 30, 25] )
```

```python
# Realizamos un grafico de dispersi贸n ( Scatter-Plot )
plt.figure( figsize=(16,4) )                           # Indicamos el tama帽o de la gr谩fica
plt.title("Scatter Plot")                              # A帽adimos el titulo
plt.scatter( x, y , label = "Etiqueta de la serie" )   # A帽adimos la 'capa' de puntos 
#plt.scatter( x, y-10 , label = "Etiqueta de la serie" )   # A帽adimos la 'capa' de puntos 
plt.xlabel("Eje X")                                    # Modificamos la etiqueta del eje X
plt.ylabel("Eje Y")                                    # Modificamos la etiqueta del eje Y
plt.grid()                                             # A帽adimos una cuadricula/rendija 
plt.legend()                                           # A帽adimos un cuadro con la leyenda
plt.show()                                             # Indicamos que queremos mostrar la gr谩fica
```

<div class="alert alert-warning">
Recordemos que 

$$S_{xx} = \sum_{i=1}^{n} (X - \bar{X} )^2 \qquad S_{xy} = \sum_{i=1}^{n} (X - \bar{X} ) \cdot (Y - \bar{Y}) \qquad S_{yy} = \sum_{i=1}^{n} (Y - \bar{Y} )^2$$
</div>

```python
# Realizamos algunos c谩lculos
Sxx = np.sum( np.power( x - np.mean(x), 2) )
Syy = np.sum( np.power( x - np.mean(x), 2) )
Sxy = np.sum( ( x - np.mean(x) ) * ( y - np.mean(y) ) )

print( "Sxx: ", Sxx )
print( "Syy: ", Syy )
print( "Sxy: ", Sxy )
```

Si consideramos el modelo de regresi贸n lineal simple con intercepto, tenemos los siguientes estimadores para las betas

$$\hat{\beta_1} = \frac{S_{xy}}{S_{xx}} \qquad \hat{\beta_0} = \bar{Y} - \hat{\beta_1} \cdot \bar{X}$$

```python
# Estimamos los par谩metros ( para el modelo con intercepto ) y la estimaci贸n de Y
b1    = Sxy / Sxx
b0    = np.mean(y) - b1 * np.mean(x)
y_est = b0 + b1 * x

print("b0: ", b0 )
print("b1: ", b1 )
```

```python
# Realizamos un grafico de dispersi贸n ( Scatter-Plot )
plt.figure( figsize=(16,4) )                           # Indicamos el tama帽o de la gr谩fica
plt.title("Scatter Plot")                              # A帽adimos el titulo
plt.scatter( x, y , label = "Etiqueta de la serie")    # A帽adimos la 'capa' de puntos 
plt.plot( x, y_est, label = "Regresi贸n", color="red")  # A帽adimos la 'capa' de una linea
plt.xlabel("Eje X")                                    # Modificamos la etiqueta del eje X
plt.ylabel("Eje Y")                                    # Modificamos la etiqueta del eje Y
plt.grid()                                             # A帽adimos una cuadricula/rendija 
plt.legend()                                           # A帽adimos un cuadro con la leyenda
plt.show()                                             # Indicamos que queremos mostrar la gr谩fica
```

```python
# Calculamos algunas metricas para ver que tan bueno es el modelo lineal
MAE = np.mean( np.abs(   y     - y_est      )    )
MSE = np.mean( np.power( y     - y_est      , 2) ) # Error Cuadratico Medio     ( MSE = Mean Square Error )
SCT = np.sum(  np.power( y     - np.mean(y) , 2) ) # Suma de cuadrados totales
SCR = np.sum(  np.power( y_est - np.mean(y) , 2) ) # Suma de cuadrados de la regresi贸n
SCE = np.sum(  np.power( y     - y_est      , 2) ) # Suma de cuadrados del error
R2  = SCR / SCT                                    # R cuadrada

print("MAE: ", MAE)
print("MSE: ", MSE)
print("SCT: ", SCT)
print("SCR: ", SCR)
print("SCE: ", SCE)
print("R2:  ", R2)
```

<div class="alert alert-success">
<center><h3>Con la biblioteca StatsModels</h3></center>
</div>

Esta biblioteca puede llegar a mostrar resultados parecidos a como lo har铆a R. 

Por lo que aquellos que est茅n acostumbrados a su manejo, puede ser una buena opci贸n, aunque no la mejor.

```python
import statsmodels.api as sm    # Biblioteca para realizar algunos modelos estadisticos
sm.OLS(y, x).fit().params       # Veamos los parametros que estima este modelo

# Notemos que por defecto, unicamente estima b1 (modelo sin intercepto)
```

```python
# La funci贸n sm.OLS(y, x) espera la matriz de dise帽o.
# Por lo que si no se incluye la columna de unos (1), no estimar谩 el incercepto.
sm.add_constant(x)              # Matriz de dise帽o con intercepto
```

```python
# Calculemos nuevamente los parametros de la regresi贸n con intercepto. 
parametros = sm.OLS(y, sm.add_constant(x) ).fit().params
print("Par谩metros: ", parametros)
```

```python
# Una manera de ver un resumen de todo el modelo es la siguiente
print( sm.OLS(y, sm.add_constant(x) ).fit().summary() )
```

<b>NOTA:</b> 
    
Sin embargo, en la mayor铆a de las situaciones trabajaremos con un DataFrame en lugar de Arrays.

Es por eso, que mostramos otra manera de realizar una regresi贸n lineal con _StatsModels_.

Adem谩s, podemos utilizar formulas como se har铆a en R.

Veamos que hay que acceder a otro modulo de la biblioteca _StatsModels_.

</div>

```python
import pandas as pd                     # Biblioteca para trabajar con DataFrames (tablas de datos)
import statsmodels.formula.api as smf   # Biblioteca para realizar algunos modelos estadisticos
```

```python
# En este caso, crearemos un DataFrame para mostrar otra forma de hacerlo
tabla_datos = pd.DataFrame( { 'y' : y, 'x' : x } )
tabla_datos.head()
```

```python
print( smf.OLS.from_formula("y ~ 0 + x", data = tabla_datos).fit().summary() )
```

```python
from sklearn import linear_model  # Biblioteca por excelencia para Machine Learning en Python

# Antes de crear nuestro modelo de regresi贸n linea, es importante verificar los par谩metros por defecto
linear_model.LinearRegression()
```

```python
modelo = linear_model.LinearRegression(fit_intercept=True)
modelo.fit( x[:, np.newaxis ] , y )
modelo.
```

```python
# Ajustemos el modelo y veamos lo que nos regresa
modelo = linear_model.LinearRegression(fit_intercept=True).fit( x[:, np.newaxis ] , y )
print( "Intercepto:   " , modelo.intercept_ )
print( "Coeficientes: " , modelo.coef_ )
print( "R2:           " , modelo.score( x[:, np.newaxis ] , y ) )
```

```python
# Si deseamos tener m谩s m茅tricas para evaluar nuestro modelo, podemos hacer uso del siguiente modulo de scikit-learn
from sklearn import metrics

# primero debemos estimar el valor de 'y' bajo nuestro modelo y luego compararlo con las 'y' reales
y_estimada = modelo.predict( x[:, np.newaxis ] )

print("MAE: ", metrics.mean_absolute_error( y, y_estimada ) )
print("MSE: ", metrics.mean_squared_error(  y, y_estimada ) )
print("R2:  ", metrics.r2_score(            y, y_estimada ) )
```
